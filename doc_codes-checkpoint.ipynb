```python
#In order to calculate KL Values and Wesserstein Distance, we need to compare the actual distributions from the random ones. Calculating uniform vote share
#to depict the random distribution since if voters are voting randomly, we expect all candidates to get uniform vote shares over repeated iterations.
df['uniform_vote_share'] = df['remaining_vote_share']/df['remaining_cand']
df['p'] = df['new_vshare_proprtn']
df['q'] = df['uniform_vote_share']
```

```python
#calculating the variables required for KL Values
import numpy as np
df['pbyq'] = ""
df["index"] = df.index
df['pbyq'] = df['index'].apply(lambda x: df.p.iloc[x]/df.q.iloc[x] if df.remaining_cand.iloc[x]>1 else 1)
df['ln(pbyq)'] = np.log(df['pbyq'])
df['p.ln(pbyq)'] = df['ln(pbyq)']*df['p']
```

```python
#In order to determine whether vote share in one row is similar to the vote share in the next row. This will help us distinguish between elections
df['match'] = df.uniform_vote_share.eq(df.uniform_vote_share.shift(periods = 1))
```

```python
#want to determine those constituencies that have the same uniform vote share in consecutive years.
df['check'] = df['index'].apply(lambda i: 'Yes' if ((df['Year'].iloc[i]!=df['Year'].iloc[i-1])&(df['uniform_vote_share'].iloc[i]==df['uniform_vote_share'].iloc[i-1])) else 'oops')
```

```python
#4 constituencies for which consecutive year had same uniform vote share. I have corrected these const here.
df['match'].iloc[22586] = False
df['match'].iloc[26906] = False
df['match'].iloc[30962] = False
df['match'].iloc[41694] = False
#The aforementioned const were 1|10|179, 1|15|16, 1|19|42, 1|23|16
```

```python
df['final_val'] = df.groupby((~df.match).cumsum())['p.ln(pbyq)'].cumsum()
```

```python
df.iloc[-10:,-17:]
```

```python
#making calculations for KL values
import numpy as np
df['qbyp'] = df['q']/df['p']
df['qbyp'] = df['index'].apply(lambda x: df.q.iloc[x]/df.p.iloc[x] if df.remaining_cand.iloc[x]>1 else 1)
df['ln(qbyp)'] = np.log(df['qbyp'])
df['q.ln(qbyp)'] = df['ln(qbyp)']*df['q']
```

```python
print(df[df['remaining_cand']==1].shape)
print(df[df['remaining_cand']==1].qbyp.sum())
#since both are equal, it means that rows values with remain_cand = 1 will now have the correct values of log and other columns. This value is 314
```

```python
df['final_val_2'] = df.groupby((~df.match).cumsum())['q.ln(qbyp)'].cumsum()
```

```python
#Apart from getting correct values per const, I have corrected the last const as well. 
import numpy as np
df['KL_1'] = np.where(df['match'].shift(-1) == False, df['final_val'], '')
df['KL_2'] = np.where(df['match'].shift(-1) == False, df['final_val_2'], '')
df['KL_1'].iloc[106401] = df['final_val'].iloc[106401]
df['KL_2'].iloc[106401] = df['final_val_2'].iloc[106401]
```
