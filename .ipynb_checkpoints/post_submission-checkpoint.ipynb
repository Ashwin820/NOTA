```python
import pandas as pd
import numpy as np
import warnings as wn
import seaborn as sns
from matplotlib import pyplot as plt
wn.filterwarnings("ignore")
```

```python
#Reading data. The csv file, cleanest_data.csv is available in the repostiory. 
data = pd.read_csv(r"D:\MSE\Dissertation\NOTA\Data\codes_documented\ancillary_files\cleanest_data.csv")
print(data.shape)
data.head()
```

```python
#Keeping only Independent candidates and NOTA for analysis, removing all other candidates. Calling this dataframe 'df'
df = data[(data['Party']=='IND')|(data['Party']=='NOTA')]
print(df.shape)
df.head()
```

```python
#Including NOTA as one of the nominated candidates in post NOTA period.
df.reset_index(inplace = True)
df['index'] = df.index
df['n_cand_nota'] = df['index'].apply(lambda i: df['N_Cand'].iloc[i]+1 if df.pre_post_nota_ind.iloc[i]==1 else df.N_Cand.iloc[i])
```

```python
#QC
df[['n_cand_nota','N_Cand','pre_post_nota_ind']].head()
#pass
```

```python
#QC
df[['n_cand_nota','N_Cand','pre_post_nota_ind']][df['pre_post_nota_ind']==1].head()
#pass
```

```python
df['nota_wins_ind'] = df['index'].apply(lambda i: 1 if ((df['Party'].iloc[i]=='NOTA')&(df['Position'].iloc[i]==1)) else 0)
```

```python
df[['Party','Position','nota_wins_ind','uni_const_code']][(df['Party']=='NOTA')&(df['Position']==1)]
#Meaning that there are no constituencies for which NOTA is winning. 
```

```python
#calculating the number of ind candidates in a constituency
len_pivot = pd.pivot_table(df[df['Party']=='IND'], index = ['uni_const_code', 'Year'], columns = ['Party'], 
                           values = 'n_cand_nota', aggfunc = lambda x: len(x)).reset_index()
print(len_pivot.shape)
len_pivot.head()
```

```python
#QC
df[(df['uni_const_code']=='0|10|10')&(df['Year']==2014)]
#pass
```

```python
#Using the following code, we will get the independent cand with min position and also NOTA. NOTA is a separate Party so it will not interfere in getting 
#the ind cand with min position even if it is winning a position lesser than the salient ind candidate.
df_larind_nota = df.groupby(['uni_const_code','Year','Party']).Position.min().reset_index()
print(df_larind_nota.shape)
df_larind_nota.tail()
```

```python
df_larind_nota = pd.merge(df_larind_nota, df[['uni_const_code','Year','Party','Position','Vote_Share_Percentage','pre_post_nota_ind']], 
                          on = ['uni_const_code','Year','Party','Position'], how = 'left')
```

```python
print(df_larind_nota.shape)
df_larind_nota.tail()
```

```python
df_larind_nota[df_larind_nota['uni_const_code']=='0|10|18']
```

```python
df_larind_pivot = df_larind_nota[df_larind_nota['pre_post_nota_ind']==1].pivot_table(columns = ['Party'], 
                                                                               index = ['uni_const_code', 'Year'],
                                                                               values = ['Vote_Share_Percentage']).reset_index()
```

```python
print(df_larind_pivot.shape)
df_larind_pivot.head()
```

```python
#combining no of ind candidates
df_larind_pivot = pd.merge(df_larind_pivot, len_pivot, on = ['uni_const_code','Year'], how = 'left')
print(df_larind_pivot.shape)
df_larind_pivot.head()
```

```python
df_larind_pivot.drop(columns = [('uni_const_code', ''),('Year', '')], inplace = True)
df_larind_pivot.rename(columns = {('Vote_Share_Percentage', 'IND'): 'vote_share_ind', ('Vote_Share_Percentage', 'NOTA'): 'vote_share_nota'}, inplace = True)
df_larind_pivot.dropna(inplace =True)
df_larind_pivot.rename(columns = {'IND':'n_ind_cand'}, inplace  = True)
print(df_larind_pivot.shape)
#the reason of this fall in rows is because there are many constituencies in which there is nota but no independent candidate.
#nota has all the entries, the missing entries are from the independent candidates
#for the current dataframe, this number is about 700. Suggests we should take other smaller and regional parties into account as well.
```

```python
#running a regression model to see the relationship between the largest independent vote share and nota vote share controlling for number of ind candidates
#Running a simple regression using Statsmodels library
import statsmodels.api as sm
X = df_larind_pivot[['vote_share_nota', 'n_ind_cand']]
X = sm.add_constant(X)
Y = df_larind_pivot[['vote_share_ind']]
model = sm.RLM(Y,X,  M=sm.robust.norms.AndrewWave()).fit()
print_model = model.summary()
print(print_model)
```

```python
import statsmodels.api as sm
X = df_larind_pivot[['vote_share_nota']]
X = sm.add_constant(X)
Y = df_larind_pivot[['vote_share_ind']]
model = sm.RLM(Y,X,  M=sm.robust.norms.AndrewWave()).fit()
print_model = model.summary()
print(print_model)
```
